   # Mag Rotator Data Analysis

```python
import os
import numpy as np
% pylab inline
```

```python
dirname = '/Volumes/NO NAME'
infname = os.listdir(dirname)[-2]
print(os.listdir(dirname))
infname = os.path.join(dirname,infname)
print(infname)
```

```python
# Read file
lines = [l.strip() for l in open(infname,'rt')]
datime = []
magy=[]
rpmset=[]
rpmnow=[]
speed=[]
for l in lines[7000:7200]:
    if(len(l)>10):
        lsplit = l.split('\t')
        magy.append(float(lsplit[5]))
        rpmset.append(float(lsplit[2]))
        rpmnow.append(float(lsplit[3]))
        speed.append(float(lsplit[1]))
print(len(magy))
print(lines[0])
print(lines[1])
```

```python
figure(figsize=(24,6))
plot(magy)
plot(magy,'rd')
title('MagY')
figure(figsize=(24,6))
#plot(rpmset)
plot(rpmnow)
title('RPMset/now')
figure(figsize=(24,6))
plot(speed)
title('Speed')
```

```python
# search for maxima in magy
maxinds = []
for i in range(2,len(magy)):
    if(magy[i-1]>magy[i-2] and magy[i-1]>=magy[i]): maxinds.append(i)
        
maxinds = np.array(maxinds)
indelt = maxinds[1:]-maxinds[:-1]
plot(indelt)
```
